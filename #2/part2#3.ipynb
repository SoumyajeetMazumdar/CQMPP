```python
from scipy import integrate
import numpy as np
from matplotlib import pyplot as plt


def f(x,a):
    return 1/(x**2 + a**2)


def simpsons(a, b, N ):
    x = []
    fx = []
    
    h = (b - a) / N
    
    for i in range(0, N+1): # creating the x array
        x.append(a+i*h)
        fx.append(f(x[i]))
        
    simpson = 0
    i = 0
    while i<= N:
        if i == 0 or i == N:
            simpson+= fx[i]
        elif i % 2 != 0:
            simpson+= 4 * fx[i]
        else:
            simpson+= 2 * fx[i]
        i+= 1
        
    simpson = simpson * (h / 3)
    return simpson






# checking the graph
a = 2 # given

x = np.linspace(0,50,100)
y = f(x,a)
plt.plot(x,y)
plt.grid(True)

# since the fn already goes to 0 at large values, we can try direct integration replacing the infinite limit using a large value.
integral, _ = integrate.quad(f, 0, 1000, args=(a,))
print(f"I = {integral}")

# value close but but not actually pi/2a
def fx(t, a):
    return 1 / (a**2 * (1 - t)**2 + t**2)

result = simpsons(0,1,10000)
print(f"The integral is approximately: {result}")

```
