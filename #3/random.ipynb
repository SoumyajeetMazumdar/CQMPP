```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.misc import derivative

# Define the function f(x) = e^x
f = lambda x: np.exp(x)

# Define x values (numpy array)
x_vals = np.linspace(0, 2, 100)

# Compute the derivative of f(x) = e^x at each x in x_vals
f_prime_analytical = [derivative(f, x, dx=1e-6, n=1) for x in x_vals]

# Compute the exact derivative analytically (for comparison)
f_prime_exact = np.exp(x_vals)

# Plot the numerical derivative and exact derivative
plt.figure(figsize=(10, 6))
plt.plot(x_vals, f_prime_analytical, label="Numerical Derivative (using scipy.derivative)", linestyle='--')
plt.plot(x_vals, f_prime_exact, label="Exact Derivative (e^x)", linestyle='-')
plt.xlabel("x")
plt.ylabel("Derivative")
plt.title("Comparison of Numerical and Exact Derivative of $e^x$")
plt.legend()
plt.grid()
plt.show()
```
