```python
import pandas as pd
from matplotlib import pyplot as plt

# first seven wkb and tise energies

tise_energies=[-0.89640318, -0.71065746, -0.55142925, -0.41704123, -0.30572899,
       -0.21563389, -0.14479584, -0.09114665, -0.05250433, -0.02656973]

wkb_energies = [
    -0.94744497,  # Energy at level 0
    -0.84765555,  # Energy at level 1
    -0.75479743,  # Energy at level 2
    -0.66867222,  # Energy at level 3
    -0.58907396,  # Energy at level 4
    -0.51579425,  # Energy at level 5
    -0.44861756,  # Energy at level 6
    -0.3873236,   # Energy at level 7
    -0.33168588,  # Energy at level 8
    -0.28147264   # Energy at level 9
]

# Compute relative errors and absolute errors
absolute_errors = [abs(tise - wkb) for tise, wkb in zip(tise_energies, wkb_energies)]
relative_errors = [abs(tise - wkb) / abs(tise) for tise, wkb in zip(tise_energies, wkb_energies)]

# Present the results
import pandas as pd
error_df = pd.DataFrame({
    "TISE Energies": tise_energies,
    "WKB Energies": wkb_energies,
    "Absolute Errors": absolute_errors
})
error_df
```

In order to compute how fast the WKB levels become accurate in comparison to the TISE levels, we will compute the relative errors assuming the TISE levels are the actual energy levels.

The decrease in the relative errors will tell us how quickly the WKB levels converge on to the actual TISE levels.

```python
# Quantum levels (0, 1, 2, ...)
levels = list(range(len(relative_errors)))

# Plot absolute errors
plt.plot(levels, absolute_errors, "o-", label="Absolute Error", color="red")
plt.xlabel("Quantum Level (k)", fontsize=12)
plt.ylabel("Absolute Error", fontsize=12)
plt.title("Absolute Errors", fontsize=14)
plt.legend(fontsize=10)
plt.grid(True)
# Adjust layout
plt.tight_layout()
plt.show()
```
