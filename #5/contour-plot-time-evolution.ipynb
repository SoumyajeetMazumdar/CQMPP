```python
import numpy as np
import matplotlib.pyplot as plt

# Grid Setup
x = np.linspace(-30, 30, 6001)  # Spatial grid
t = np.linspace(0, 15, 1501)    # Time grid

# Load the precomputed wave function data (replace with your actual data file)
wavefunction_data = np.load('psi.npy')  # Example: 6001x1501 matrix
prob_density = np.abs(wavefunction_data)**2  # Calculate probability density

# Plotting
plt.figure(figsize=(10, 6))
plt.imshow(prob_density, extent=[x[0], x[-1], t[0], t[-1]], origin='lower',
           aspect='auto', cmap='jet', vmax=0.5)

# Correcting Axes Labels
plt.colorbar(label=r'$|\psi(x,t)|^2$')
plt.xlabel('Position x')
plt.ylabel('Time t')

# Overlay the barrier
plt.axvline(x=20, color='r', linestyle='-', label='Barrier')  # Barrier at x = 20
plt.text(21, 10, 'Barrier', color='red')

# Title and Legend
plt.title('Time Evolution of Wave Packet')
plt.legend()

plt.show()
```
