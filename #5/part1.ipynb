```python
import numpy as np
import matplotlib.pyplot as plt


# defining the gaussian parameters given initially
k0 = 4
sigma0 = 2
x0 = -10

# defining spatial grid
dx = 0.01         
x_min, x_max = -30, 30  
x = np.arange(x_min, x_max + dx, dx)

# Initial Gaussian wave packet (unnormalized)
psi_initial = np.exp(1j * k0 * x) * np.exp(-(x - x0)**2 / (2 * sigma0**2))

# Calculate the probability density array
prob_density = np.abs(psi_initial)**2

# Plot the initial Gaussian wave packet
plt.figure(figsize=(8, 5))
plt.plot(x, psi_initial, label='$|\psi(x, 0)|^2$', linewidth=2)
plt.title('Initial Gaussian Wave Packet')
plt.xlabel('x')
plt.ylabel('$|\psi(x, 0)|^2$')
plt.grid(True)
plt.legend()

# Plot the real and imaginary parts of the initial wavefunction
# plt.figure(figsize=(8, 5))
# plt.plot(x, np.real(psi_initial), label='Re($\psi(x, 0)$)', linewidth=2)
# plt.plot(x, np.imag(psi_initial), label='Im($\psi(x, 0)$)', linewidth=2)
# plt.title('Initial Gaussian Wave Packet')
# plt.xlabel('x')
# plt.ylabel('$\psi(x, 0)$')
# plt.grid(True)
# plt.legend()

```

```python
# Compute the normalization factor
norm_factor = np.sqrt(np.sum(prob_density) * dx)

# Normalize the wavefunction
psi_normalized = psi_initial / norm_factor

# Verify the norm after normalization
norm_check = np.sum(np.abs(psi_normalized)**2) * dx

norm_factor, norm_check
```

```python
# defining the spatial grid here
xs = -30
xe = 30
dx = 0.01
x = np.arange(xs, xe + dx, dx) 

def Vx(x, xb, Vb): # define potential function, with a step barrier at xb = 20 of height 50
    pot = np.zeros(len(x))
    for i in range(0, len(x)):
        if x[i]<=xb:
            pot[i]=0
        else:
            pot[i]=Vb
    return pot

V = lambda x: Vx(x, -10, 50)

plt.plot(x, V)
```
