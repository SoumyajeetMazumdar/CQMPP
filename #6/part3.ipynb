```python
import numpy as np
from matplotlib import pyplot as plt

counter_arr = [x for x in range(1, 1150+1)] # creating the iteration array for the x-axis

residuums_sor = np.load("residuum-arr-sor.npy")
residuums_gs = np.load("residuum-arr-gs.npy")


plt.plot(counter_arr, residuums_sor, label="SOR, p = 1.99")
plt.plot(counter_arr, residuums_gs, label="GS, p = 1.0")
plt.legend()
plt.grid()
plt.yscale("log")
plt.title("Comparison of residuum convergence")
plt.xlabel("Iterations")
plt.ylabel("Residuum")
```
