```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
N = 200  # Number of grid points
dr = 0.2  # Grid spacing
r = np.linspace(dr, N * dr, N)  # Radial grid (r != 0)

# Potential: V(r) = -1/r
V = -1 / r

# Laplacian matrix (-1/2 d^2/dr^2)
laplacian = -0.5 / dr**2 * (np.diag(-2 * np.ones(N)) + 
                            np.diag(np.ones(N-1), k=1) +
                            np.diag(np.ones(N-1), k=-1))

# Hamiltonian matrix
H = laplacian + np.diag(V)

# Solve eigenvalue problem
eigenvalues, eigenvectors = np.linalg.eigh(H)

# Ground state energy and wavefunction
E_ground = eigenvalues[0]
psi_ground = eigenvectors[:, 0]

# Normalize the wavefunction
psi_ground_normalized = psi_ground / np.sqrt(np.sum(psi_ground**2) * dr)

# Output results
print(f"Ground state energy: {E_ground:.6f}")
print(f"Expected energy (exact): -0.500000")

# Plot the ground-state wavefunction
plt.plot(r, psi_ground_normalized, label="Ground State Wavefunction")
plt.xlabel("r")
plt.ylabel("Ïˆ(r)")
plt.title("Ground State Wavefunction of Hydrogen Atom")
plt.legend()
plt.show()
```
